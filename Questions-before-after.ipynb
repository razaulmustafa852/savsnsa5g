```python
def replace_nan_and_fill_mean(df, columns):

    for col in columns:
        # Step 1: Replace '-' and empty strings with NaN
        df[col] = df[col].replace(['-', ''], np.nan)
        
        # Step 2: Convert the column to numeric (if needed)
        df[col] = pd.to_numeric(df[col], errors='coerce')
        
        # Step 3: Replace NaN with the mean of the column
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)
    
    return df
```

```python
columns_to_process = ['Qual']
ndb = replace_nan_and_fill_mean(df, columns_to_process)
nda = replace_nan_and_fill_mean(df_after, columns_to_process)
```

```python
sns.kdeplot(data=ndb, x="Qual",label= "Before 10/s")
sns.kdeplot(data=nda, x="Qual",label= "After 10/s")
plt.legend()
plt.xlabel("RSRQ")
plt.savefig("noms/rsrq-sa-ex-mobility.pdf")
```
